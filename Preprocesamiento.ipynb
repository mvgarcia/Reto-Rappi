```python
#Funcion de transformaci贸n

def img_transform(img, n):
  if n==0:  # Escala de Grises
    img_aug = imagus.grayscale(img)    
  elif n==1:  # Brillo 
    a = np.random.uniform(0.5, 1.5)
    img_aug = imagus.brightness(img, factor=a)   
  elif n==2:  # Saturaci贸n
    a = np.random.uniform(3, 5)
    img_aug = imagus.saturation(img, factor=a)
  elif n==3:  # Ruido
    a = np.random.uniform(0.02, 0.05)
    img_aug = imagus.random_noise(img, mean=0, var=a)
  elif n==4:  # Contraste
    a = np.random.uniform(0.25, 0.5)
    img_aug = imagus.contrast(img, factor=a)
#   elif n==5:  # Perspectiva
#     a = np.random.uniform(50, 100)
#     img_aug = imagus.perspective_transform(img, sigma=a)
#   elif n==6:  # Cro
#     img_aug = imagus.crop(img, x1=0, y1=0.1, x2=0.9, y2=1)
#   elif n==7:  # Meme
#     img_aug = imagus.meme_format(img, text='LOL')
#   else: # Perspectiva + Meme
#     img_aug = imagus.meme_format(img, text='LMAO')
#     img_aug = imagus.saturation(img_aug, factor=5)
  return img_aug
```
